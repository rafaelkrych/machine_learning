Basics of the API:

Most commonly, the stepis in using the Scikit-Learn estimator API are as follows (we will step throuh a handful of detailed examplels in the sections that follow):

    3. Arrange data into a features matrix and target vector
    4. Fit the model to your data by calling the fit() method of the model instance
    5. Apply the model to new data:
        - For supervised learning, often we predict labels for unknow data using the predict() method
        - For unsupervised learning, we often transform or infer properties of the data using the transform() or predict() method

## Robust Regression

Outlier Demo: http://digitalfirst.bfwpub.com/stats_applet/stats_applet_5_correg.html

```python
df.head()
```

### RANdom SAmple Consensus (RANSAC) Algorithm

link = http://scikit-learn.org/stable/modules/linear_model.html#ransac-regression

Each iteration performs the following steps:  
<ol>
<li>Select <code>min_samples</code> random saples from the original data and check whether the set of data is valid (see <code>is_data_valid</code>).</li>
<li>Fit a model to the random subset (<code>base_estimator.fit</code>) and check whether the estimated model is valid (see <code>is_model_valid</code>)</li>
<li>Classify all data as inliers or outliers by calculating the residuals to the estimated model (<code>base_estimator.predict(X) - y</code>) - all data samples with absolute residuals smaller than the <code>residual_threshold</code> are considered as inliers</li>
<li>Save fitted model as best model if number of inlier samples is maximal. In case the current estimated model has the same number of inliers, it is only considered as the best model if it has better score</li>
</ol>

```python
X = df['RM'].values.reshape(-1,1)
y = df['MEDV'].values
```

```python
from sklearn.linear_model import RANSACRegressor
```

```python
ransac = RANSACRegressor()
```

```python
ransac.fit(X,y)
```

```python
inlier_mask = ransac.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)
```

Creating range of numbers, which will be length of the line as plot

```python
np.arange(3,10,1)
```

```python
line_X = np.arange(3,10,1)
line_y_ransac = ransac.predict(line_X.reshape(-1,1))
```

```python
sns.set(style='darkgrid', context='notebook')
plt.figure(figsize=(12,8))
plt.scatter(X[inlier_mask],y[inlier_mask],
            c='blue', marker='o', label='Inliers')
plt.scatter(X[outlier_mask],y[outlier_mask],
            c='brown', marker='s', label='Outliers')
plt.plot(line_X,line_y_ransac, color='red')
plt.xlabel('Average number of roms per dwelling')
plt.ylabel("Median value of owner-occupied homes in $1000's")
plt.legend(loc='upper left')
plt.show()
```

```python
ransac.estimator_.coef_
```

```python
ransac.estimator_.intercept_
```

```python
X = df['LSTAT'].values.reshape(-1,1)
y = df['MEDV'].values
ransac.fit(X,y)
inlier_mask = ransac.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)
line_X = np.arange(0,40,1)
line_y_ransac = ransac.predict(line_X.reshape(-1,1))
```

### Let's see how it looks with our Robust Regression

```python
sns.set(style='darkgrid', context='notebook')
plt.figure(figsize=(12,8))
plt.scatter(X[inlier_mask],y[inlier_mask],
            c='blue', marker='o', label='Inliers')
plt.scatter(X[outlier_mask],y[outlier_mask],
            c='brown', marker='s', label='Outliers')
plt.plot(line_X, line_y_ransac, color='red')
plt.xlabel('% lower status of the population')
plt.ylabel("Median value of owner-occupied homes in $1000's")
plt.legend(loc='upper right')
plt.show()
```

Prediction for 5% lower status population with Robust Regression

```python
ransac.predict(np.array([5]).reshape(-1,1))
```

### And now example how it looks with Linear Regression model

```python
ml_3 = LinearRegression()
ml_3.fit(X,y)
prediction = ml_3.predict(line_X.reshape(-1,1))
sns.set(style='darkgrid', context='notebook')
plt.figure(figsize=(12,8))
plt.scatter(X,y, c='blue', marker='o', label='Inliers and Outliers')
plt.plot(line_X,prediction, color='red')
plt.xlabel('% lower status of the population')
plt.ylabel("Median value of owner-occupied homes in $1000's")
plt.legend(loc='upper right')
plt.show()
```

Prediction for 5% lower status of population with Linear Regression

```python
ml_3.predict(np.array([5]).reshape(1,-1))
```

## Performance Evaluation of Regression Model

```python
from sklearn.model_selection import train_test_split
```

```python
df.head()
```

```python
#X = df['LSTAT'].values.reshape(-1,1)
X = df.iloc[:, :-1].values # it's the same as df.values, but without last column (MEDV)
```

```python
y = df['MEDV'].values
```

```python
X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_train_pred = lr.predict(X_train)
```

```python
y_test_pred = lr.predict(X_test)
```

***

## Method 1: Residual Analysis

```python
plt.figure(figsize=(12,8))
plt.scatter(y_train_pred, y_train_pred - y_train, c='blue', marker='o', label='Training data')
plt.scatter(y_test_pred, y_test_pred - y_test, c='orange', marker='*', label='Test data')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.hlines(y=0, xmin=-10, xmax=50, lw=2, color='k')
plt.xlim([-10, 50])
plt.legend(loc='upper left')
plt.show
```

***

## Method 2: Mean Square Error (MSE)

<ul>
    <li>The average value of the Sums of Squared Error cost function</li>
    <li>Useful for comparing different regression models</li>
    <li>For tuning parameters via a grid search and cross-validation</li>
</ul>

Lower number = better

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_train, y_train_pred)
```

```python
mean_squared_error(y_test, y_test_pred)
```

***

## Method 3: Coefficient of Determination, _R_<sup>2</sup>

SSE: Sum of squared errors  
SST: Total sum of squares

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_train, y_train_pred)
```

```python
r2_score(y_test, y_test_pred)
```

## What does a Near Perfect Model looks like?

```python
generate_random = np.random.RandomState(0)
x = 10 * generate_random.rand(1000)
y = 3 * x + np.random.randn(1000)   # 3*x + some noise
plt.figure(figsize=(10,8))
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True) #fit_intercept=False sets the y-intercept to 0. If fit_intercept=True, the y-intercept will be determined by the line of best fit.
model.fit(X_train.reshape(-1,1), y_train)

y_train_pred = model.predict(X_train.reshape(-1,1))
y_test_pred = model.predict(X_test.reshape(-1,1))
```

## Method 1: Residual Analysis

```python
plt.figure(figsize=(12,8))
plt.scatter(y_train_pred, y_train_pred - y_train, c='blue', marker='o', label='Training data')
plt.scatter(y_test_pred, y_test_pred - y_test, c='orange', marker='*', label='Test data')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.legend(loc='upper left')
plt.hlines(y=0, xmin=-3, xmax=33, lw=2, color='k')
plt.xlim([-5, 35])
plt.ylim([-25, 15])
plt.show()
```

## Method 2: Mean Squared Error (MSE)

```python
mean_squared_error(y_train, y_train_pred)
```

```python
mean_squared_error(y_test, y_test_pred)
```

## Method 3: Coefficient of Determination, R<sup>2</sup>

```python
r2_score(y_train, y_train_pred)
```

```python
r2_score(y_test, y_test_pred)
```
